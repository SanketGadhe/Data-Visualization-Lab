```python
# Box plot
import matplotlib.pyplot as plt
import numpy as np
# taking sample data
data=np.array([12,14,18,16,15,10,17,20,23,19])
plt.boxplot(data)
plt.show()
```

```python
# pie chart
import matplotlib.pyplot as plt
data=[20,30,70,50]
label=['A','B','C','D']
plt.pie(data,labels=label)
plt.show()
```

```python
# scatter plot
import matplotlib.pyplot as plt
x=[5,7,9,11,13,17,19,21,23,25,27]
y=[80,96,65,76,109,56,89,134,98,22,103]
plt.scatter(x,y,c="blue")
plt.show()
```

```python
# Heat map
import seaborn as sns
import matplotlib.pyplot as plt
w1=[24,20,21,19,17]
w2=[16,19,23,15,20]
w3=[22,23,25,21,24]
w4=[14,12,15,11,10]
data=[w1,w2,w3,w4]
sns.heatmap(data)
plt.show()
```

```python
# Histogram
import matplotlib.pyplot as plt
x=[10,20,30,40,50,60,12,14,53,54,25,22,18,19,35,60,22,48,32,12,19,44,49,43,23]
plt.hist(x)
plt.show()
```
